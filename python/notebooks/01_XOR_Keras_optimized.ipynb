# 01 - optimized XOR model with Keras

```python
import sys
import numpy as np
import tensorflow as tf
import keras.backend as K
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import sigmoid
from keras.activations import relu
from keras.activations import tanh
from keras.losses import MSE
from keras.losses import mean_squared_error
from keras.losses import binary_crossentropy
from keras.optimizers import SGD, Adam
from keras.metrics import binary_accuracy

from tensorflow.python.framework.graph_util import convert_variables_to_constants

import keras
print keras.__name__, keras.__version__
print tf.__name__, tf.__version__
```

Note: Keras should be Version 2.2.2 and TensorFlow 1.12.0

#### Tensorflow and Keras session

```python
sess=tf.Session()
K.set_session(sess)
```

#### training and test data

```python
training_data = np.array([[0,0], [0,1], [1,0], [1,1]])
target_data   = np.array([  [0],   [1],   [1],   [0]])
```

#### XOR model optimized

- activation [Relu](https://en.wikipedia.org/wiki/Relu)
![Rectifier (neural networks)](https://upload.wikimedia.org/wikipedia/commons/thumb/f/fe/Activation_rectified_linear.svg/320px-Activation_rectified_linear.svg.png) (credit: https://commons.wikimedia.org/wiki/File:Activation_rectified_linear.svg)
- another layer
- loss binary_crossentropy
- optimizer Adam

```python
model = Sequential()
# model.add(Dense(5, input_dim=2, activation='sigmoid', name='hiddenlayer'))
model.add(Dense(5, input_dim=2, activation='relu', name='hiddenlayer'))

# another layer
model.add(Dense(10, activation='relu', name='anotherlayer'))

model.add(Dense(1, activation='sigmoid', name='outputlayer'))
```

```python
# model.compile(loss='mean_squared_error', optimizer=SGD(lr=1), metrics=['accuracy'])
model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.1), metrics=['accuracy'])
```

```python
from keras.callbacks import TensorBoard
from tensorflow.contrib.tensorboard.plugins import projector
tensorboard = TensorBoard(log_dir='./logs/xor_keras_optimized', histogram_freq=0,
                          write_graph=True, write_images=False)
```

#### start training

```python
epochs = 1000
model.fit(training_data, target_data, epochs=epochs, callbacks=[tensorboard])
```

#### result

```python
print "loss:", model.evaluate(x=training_data, y=target_data, verbose=0)
print ""
print model.predict(training_data)
print ""
print model.predict(training_data).round()
```
