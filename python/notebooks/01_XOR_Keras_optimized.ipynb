# 01 - optimized XOR model with Keras

````python
import sys
import numpy as np
import tensorflow as tf
import keras.backend as K
from keras.models import Sequential
from keras.layers import Dense
from keras.activations import sigmoid
from keras.activations import relu
from keras.activations import tanh
from keras.losses import MSE
from keras.losses import mean_squared_error
from keras.losses import binary_crossentropy
from keras.optimizers import SGD, Adam
from keras.metrics import binary_accuracy

from tensorflow.python.framework.graph_util import convert_variables_to_constants

import keras
print keras.__name__, keras.__version__
print tf.__name__, tf.__version__
````

Keras sollte die Version 2.2.2 und TensorFlow 1.12.0 haben.

#### Tensorflow and Keras session

````python
sess=tf.Session()
K.set_session(sess)
````

#### training and test data

````python
training_data = np.array([[0,0], [0,1], [1,0], [1,1]])
target_data   = np.array([  [0],   [1],   [1],   [0]])
````

#### XOR model

![Neural network](https://upload.wikimedia.org/wikipedia/commons/thumb/3/3d/Neural_network.svg/320px-Neural_network.svg.png) (credit: https://commons.wikimedia.org/wiki/File:Neural_network.svg)

````python
model = Sequential()
# model.add(Dense(5, input_dim=2, activation='sigmoid', name='inputlayer'))
model.add(Dense(5, input_dim=2, activation='relu', name='inputlayer'))

# another layer
model.add(Dense(10, activation='relu', name='anotherlayer'))

model.add(Dense(1, activation='sigmoid', name='outputlayer'))

# model.compile(loss='mean_squared_error', optimizer=SGD(lr=1), metrics=['accuracy'])
model.compile(loss='binary_crossentropy', optimizer=Adam(lr=0.1), metrics=['accuracy'])
````

input layer with hidden layer
- [dense](https://keras.io/layers/core/#dense)
- 5 hidden nodes
- 2 input values (training data)

output Layer
- 1 node

activation method: [sigmoid function](https://en.wikipedia.org/wiki/Sigmoid_function)
![sigmoid function](https://upload.wikimedia.org/wikipedia/commons/thumb/8/88/Logistic-curve.svg/640px-Logistic-curve.svg.png) (credit: https://commons.wikimedia.org/wiki/File:Logistic-curve.svg)

loss function: [mean_squared_error](https://keras.io/losses/)

optimizer: [SGD](https://keras.io/optimizers/) 
- [Stochastic gradient descent](https://en.wikipedia.org/wiki/Stochastic_gradient_descent)
- [XOR tutorial with TensorFlow](https://martin-thoma.com/tf-xor-tutorial/)

#### TensorBoard
[TensorBoard Callback](https://keras.io/callbacks/#tensorboard)

````python
from keras.callbacks import TensorBoard
from tensorflow.contrib.tensorboard.plugins import projector
tensorboard = TensorBoard(log_dir='./logs/xor_keras_optimized', histogram_freq=0,
                          write_graph=True, write_images=False)
````

#### start training

````python
epochs = 500
model.fit(training_data, target_data, epochs=epochs, callbacks=[tensorboard])
````

#### result

````python
print "loss:", model.evaluate(x=training_data, y=target_data, verbose=0)
print ""
print model.predict(training_data)
print ""
print model.predict(training_data).round()
````

## freeze model and save to PB file

````python
freeze_var_names = list(set(v.op.name for v in tf.global_variables()))
print freeze_var_names
output_names = [model.outputs[0].name.split(":")[0]]

from tensorflow.python.framework.graph_util import remove_training_nodes
sub_graph_def = remove_training_nodes(sess.graph_def)

frozen_graph = convert_variables_to_constants(sess,
                                              sub_graph_def,
                                              output_names,
                                              freeze_var_names)

graph_path = tf.train.write_graph(frozen_graph, "models", "xor_keras_optimized.pb", as_text=False)
print output_names
print('%s written' % graph_path)
````
